#### Generación de Clientes

Para cada cliente se crean los siguientes atributos:
* ID de Cliente
* Nombre y apellido
* Email
* Ciudad y Provincia
* Fecha de Registro

Para la generación de los clientes, vamos a definir las siguientes funciones:
* fecha_aleatoria --> Para generar la fechaRegistro del Cliente
* normalizar_texto --> Para normalizar mails
* generar_mail_unico --> Para garantizar mails unicos en el dataset.
* generar_clientes --> Para generar los clientes

```python
# La siguiente funcion recibe dos fechas como parámetro y lo que hace es generar 
# una fecha aleatoria que se encuentre en el rango de fechas dadas 

def fecha_aleatoria(inicio, fin):
    delta = fin - inicio
    dias = random.randint(0, delta.days)
    return inicio + timedelta(days=dias) ## timedelta suma los dias calculados a la fecha inicio
```

```python
# Set para almacenar emails ya generados (evita duplicados)
emails_existentes = set()

def normalizar_texto(texto):
    """ Esta función:
    - Convierte a minúsculas
    - Elimina tildes
    - Elimina la ñ
    - Elimina espacios
    - Elimina caracteres especiales
    Esto hace que el email sea más realista y válido.
    """
    texto = texto.lower()  # todo en minúscula
    texto = unicodedata.normalize('NFD', texto)  # Separa letras de acentos (ej: á -> a + ´)
    texto = ''.join(c for c in texto if unicodedata.category(c) != 'Mn') # Elimina los acentos (borra los caracteres de tipo "marca")
    texto = texto.replace(" ", "") # Elimina espacios
    return texto


def generar_email_unico(nombre, apellido):
    ##Genera un email único incluso si existen dos personas con el mismo nombre y apellido
    
    # Normalizamos nombre y apellido
    nombre_norm = normalizar_texto(nombre)
    apellido_norm = normalizar_texto(apellido)
    
    base = nombre_norm + apellido_norm
    dominio = random.choice(dominios)
    
    email = base + dominio
    contador = 1
    
    # Si el email ya existe, agregamos un número al final
    while email in emails_existentes:
        email = f"{base}{contador}{dominio}"
        contador += 1
    
    # Guardamos el email en el set para no repetirlo
    emails_existentes.add(email)
    
    return email
```

```python
def generar_clientes(cantidad, inicio, fin, id_actual):
    lista_clientes = []
    id = id_actual

    for _ in range(cantidad):
        nombre = random.choice(nombres)
        apellido = random.choice(apellidos)
        provincia = random.choice(list(localidades.keys()))
        ciudad = random.choice(localidades[provincia])

        fecha = fecha_aleatoria(inicio, fin)

        cliente = {
            'idCliente': id,
            'nombre': nombre,
            'apellido': apellido,
            'email': generar_email_unico(nombre, apellido),
            'fechaRegistro': fecha,
            'provincia': provincia,
            'ciudad': ciudad
        }

        lista_clientes.append(cliente)
        id += 1

    return lista_clientes, id
```

### Generación de Ventas

Para la generación de las Ventas si bien vamos a usar funciones random, utilizaremos pesos en ciertas variables para que la funcion elija según mas probabilidades o menos.

Para eso tenemos que definir, pesos en:
* Categorias --> qué categorías son las que tienen mas probabilidad de ventas
* Mes --> hay meses que tienen tendencias a tener mayores ventas, por varios factores como Navidad por ejemplo

Tambien se definirá mas pesos para los productos estrella

Para la generación de las ventas se definen varias funciones:
* ajustar_pesos --> define los pesos finales para la elección del producto 
* generar_fecha_ponderadad --> genera la fecha de la transaccion teniendo en cuenta la fecha registro del cliente elegido y la fecha final del período de análisis
* elegir_producto --> elige el producto usando los pesos ajustados de la primer funcion
* calcular_cantidad --> calcula la cantidad de producto por cada transaccion dependiendo del precio
* generar_estado_pedido --> define el estado de cada pedido segun fecha de entrega
* generarVentas --> genera cada transaccion con las funciones anteriormente nombradas y con los datos de cantidad de ventas anual, el año, los clientes, los ids y los pesos de clientes que el primer año será 1 para todos, es decir, todos tienen la misma probabilidad de ser seleccionados

```python
pesos_categoria = {
    1: 1.5,   # Celulares
    2: 1.4,   # Notebooks
    3: 0.8,   # Tablets
    4: 1.0,   # Relojes
    5: 1.6,   # Auriculares
    6: 1.2    # Parlantes
}
peso_mes = {
    1: 0.95,   # Enero - baja post fiestas
    2: 1.2,    # Inicio clases
    3: 1.2,    # Inicio clases fuerte
    4: 1,      # Mes normal
    5: 1,      # Mes normal
    6: 1.3,    # Aguinaldo
    7: 1.1,    # Vacaciones (depende rubro)
    8: 1,      # Mes normal
    9: 1.1,    # Empieza Q4
    10: 1.25,  # Día de la Madre
    11: 1.3,   # Cyber / Black Friday
    12: 1.35   # Navidad (máximo pico)
}
```

```python
def ajustar_pesos(fecha, prod):
    # Peso base categoría
    peso = pesos_categoria[prod["idCategoria"]]

    # Estacionalidad general mes
    peso *= peso_mes[fecha.month]

    # INICIO DE CLASES (Feb/Mar)
    if fecha.month in [2, 3]:
        if prod["idCategoria"] in [2, 3]:  # Notebooks y Tablets
            peso *= 1.6
            
    # AGUINALDO (Jun/Jul)
    if fecha.month in [6, 7]:
        if prod["idCategoria"] == 2:  # Notebooks
            peso *= 1.5
        if prod["precio"] > 600000:  # Gama alta
            peso *= 1.3

    # DÍA DE LA MADRE (Oct)
    if fecha.month == 10:
        if prod["idCategoria"] in [4, 5]:  # Relojes y Auriculares
            peso *= 1.6
            
    # CYBER (Nov)
    if fecha.month == 11:
        if prod["precio"] < 300000:  # Productos masivos (mas baratos - mas cantidad)
            peso *= 1.5

    # NAVIDAD (Dic)
    if fecha.month == 12:
        if prod["idCategoria"] in [4, 5]:
            peso *= 1.8
        if prod["idCategoria"] == 1 and prod["precio"] < 400000:
            peso *= 1.4
            
    # Productos estrella
    if prod["producto"] in productos_estrella:
        peso *= 1.5

    return peso
```

```python
def generar_fecha_ponderada(inicio, fin):
    
    combinaciones = []
    pesos = []
    
    for anio in range(inicio.year, fin.year + 1):
        for mes in range(1, 13):
            
            # Creamos fecha representativa del mes
            primer_dia = datetime(anio, mes, 1)
            ultimo_dia = datetime(anio, mes, calendar.monthrange(anio, mes)[1])

            if ultimo_dia >= inicio and primer_dia <= fin:
                
                combinaciones.append((anio, mes))
                peso_total = peso_mes[mes] 
                pesos.append(peso_total)
    
    # Elegimos combinación válida
    anio, mes = random.choices(combinaciones, weights=pesos)[0]
    
    # Día válido real
    ultimo_dia = calendar.monthrange(anio, mes)[1]
    dia = random.randint(1, ultimo_dia)
    
    fecha_final = datetime(anio, mes, dia)
    
    # Si el día quedó fuera del rango exacto, ajustamos
    if fecha_final < inicio:
        fecha_final = inicio
        
    if fecha_final > fin:
        fecha_final = fin
    
    return fecha_final
```

```python
def elegirProducto(f):

    pesos_productos = []

    for prod in productos:
        peso = ajustar_pesos(f, prod)
        pesos_productos.append(peso)

    producto = random.choices(productos, weights=pesos_productos)[0]
    return producto
```

```python
def calcular_cantidad(prod):
    precio = prod["precio"]

    # Económicos (menos de 200k)
    if precio < 200000:
        return random.choices([1, 2, 3], weights=[2, 4, 3])[0]

    # Intermedios (200k a 700k)
    elif precio < 700000:
        return random.choices([1, 2],weights=[5, 1])[0]

    # Premium (más de 700k)
    else:
        base = 1

    return base
```

```python
def generar_estado_pedido(fecha_entrega):
    hoy = datetime(2026, 2, 1)  # simulamos cierre de análisis
    
    if fecha_entrega < hoy:
        return random.choices(["Entregado", "Cancelado"], weights=[93, 7])[0]
    else:
        return "Pendiente"
```

```python
def generarVentas (cantidad_ventas, anio, clientes, idV, pesos_clientes):
    ventas_local = []
    inicio_anio = datetime(int(anio), 1, 1)
    fin_anio = datetime(int(anio), 12, 31)
    clientes_validos = [c for c in clientes if c['fechaRegistro'] <= fin_anio]

    pesos_validos = [pesos_clientes[clientes.index(c)] for c in clientes_validos]
    
    for i in range(1, cantidad_ventas+1):
        cliente_elegido = random.choices(clientes_validos, weights=pesos_validos)[0]
        idC = cliente_elegido['idCliente']
    
        fechaRegistro = cliente_elegido['fechaRegistro']  #tomamos la fecha de registro del cliente que debe ser anterior a la compra
        # la venta no puede ser antes del registro
        fecha_inicio_real = max(fechaRegistro, inicio_anio)
        
        fecha = generar_fecha_ponderada(fecha_inicio_real, fin_anio)
    
        producto = elegirProducto(fecha)
        idP = producto['idProducto']       #tomamos su id
        
        met = random.randint(0, len(metodosDePago)-1)     #elegimos un metodo de pago al azar
        idM = metodosDePago[met]['idPago']                #tomamos su id
        
        fecha_entrega = fecha + timedelta(days=random.randint(1, 10)) #le sumamos unos entre 1 a 10 dias a la fecha de entrega, el pedido no puede tardar mas de 10 dias
        cantidad = calcular_cantidad(producto)
        
        venta = {
            'idVenta' : idV, 
            'idCliente' : idC,
            'fechaTransaccion': fecha,
            'idProducto': idP,
            'cantidad' :  cantidad,
            'metodoPago' : idM,
            'estado' : generar_estado_pedido(fecha_entrega),
            'fechaEntrega' : fecha_entrega,
            'total': producto['precio'] * cantidad
        }
        ventas_local.append(venta)
        idV = idV+1
    return ventas_local, idV
```

```python
# Hacemos el primer llamado a la función generar ventas con la lista de ventas ya creada,
# la cantidad de ventas ya definida para 2024 y los pesos con mismo valor para todos 
ventas = []
cantidad_2024 = 750
idVenta = 1
pesos_clientes = [1] * len(clientes_2024)

ventas2024, idVenta = generarVentas(cantidad_2024, "2024", clientes_2024, idVenta, pesos_clientes)

# chequeamos el número de ventas realizada
ventas += ventas2024
print("Cantidad de Ventas 2024: ", len(ventas))
```

Ahora usaremos pandas para poder convertir ventas en una tabla y realizar lo siguiente:

* Se toma el primer 20% de los clientes y se definen como frecuentes

```python
compras_por_cliente = (
    dfVentas[dfVentas["fechaTransaccion"].dt.year == 2024]
    .groupby("idCliente")
    .size())

cantidad_frecuentes = int(len(clientes_2024) * 0.20)

clientes_frecuentes_ids = compras_por_cliente.sort_values(ascending=False) \
                                            .head(cantidad_frecuentes) \
                                            .index.tolist()
```

```python
# Definimos la total de clientes de 2025
# Sacamos la diferencia con los clientes que ya teníamos

cantClientes_2025 = 595
cantidadNuevos = cantClientes_2025 - len(clientes_2024)

# Se hace la segunda llamada a generar_clientes para generar a los nuevos de 2025

clientesNuevos, id_Clientes = generar_clientes(cantidadNuevos, datetime(2025,1,1), fecha_fin, id_Clientes)
clientes.extend(clientesNuevos)

# Definimos pesos_base que usará numpy para intentar realizar una distribución pareto de los clientes
pesos_base = np.random.pareto(a = 2.5, size=len(clientes)) + 1
```

```python
# El siguiente bucle le asigna un peso distinto a cada cliente
# dependiendo de la cantidad de compras de cada uno

for i, cliente in enumerate(clientes):
    idC = cliente["idCliente"]
    base = pesos_base[i]
    
    if idC in compras_por_cliente:
        compras_previas = compras_por_cliente[idC]
        peso = base * (1 + compras_previas * 0.3)
    else:
        peso = base
        
    pesos_clientes.append(peso)
```
